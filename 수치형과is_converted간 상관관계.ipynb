```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_train = pd.read_csv("train.csv") # 학습용 데이터
```

```python
df_train.info()
```

```python
pd.DataFrame(df_train)
```

```python
numerical_cols = df_train.select_dtypes(include=['int64', 'float64']).columns.tolist()
numerical_cols.append("is_converted")

numerical_cols
```

```python

corr = df_train[numerical_cols].corr(method='pearson')
print(corr)
```

```python
#수치형과 is_converted간 상관관계
corr_converted = corr[['is_converted']].sort_values(by='is_converted', ascending=False)
corr_converted
```
